# IoT Security & Privacy Breaches

The Internet of Things (IoT) refers to a network of physical devices, vehicles, home appliances, and other items embedded with electronics, software, sensors, and connectivity, enabling these objects to connect and exchange data.

Any Internet-enabled device is vulnerable to being hacked and misused.
An IoT security breach occurs when unauthorized access is gained to an IoT device or network, potentially allowing cybercriminals to steal data, take over devices, or launch further attacks.

```python
def extract_texts(url):
    """
    this function extracts text from all URLs with error messages delivered if there are any websites from which 
    the function fails to extract text due to any reason; it also has a time delay before extracting texts from sites
    so as to not overwhelm the kernel;
    
    # Test with a non-existent URL
    >>> extract_texts('https://this-website-does-not-exist-12345.com')
    Attempt 1 failed for https://this-website-does-not-exist-12345.com: ...
    Waiting ... seconds before retrying...
    Attempt 2 failed for https://this-website-does-not-exist-12345.com: ...
    Waiting ... seconds before retrying...
    Attempt 3 failed for https://this-website-does-not-exist-12345.com: ...
    
    # Test with Python's official website
    >>> python_text = extract_texts('https://www.python.org')
    >>> len(python_text) > 0  # Should extract some content
    True
    >>> 'Python' in python_text  # Should contain the word Python
    True
    
    # Test with a website that provides clear paragraph structure
    >>> wiki_text = extract_texts('https://en.wikipedia.org/wiki/Python_(programming_language)')
    >>> 'programming language' in wiki_text.lower()
    True
    """
    headers = {
        'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/91.0.4472.124 Safari/537.36'
    }
    
    max_retries = 3
    retry_delay = 5
    
    for attempt in range(max_retries):
        try:
            response = requests.get(url, headers=headers, timeout=30)
            response.raise_for_status()  # Raise an exception for 4XX/5XX responses
            
            soup = BeautifulSoup(response.text, "html.parser")
            # Extract all paragraph text
            paragraphs = soup.find_all("p")
            text = "\n".join([p.get_text() for p in paragraphs])
            
            return text
            
        except requests.exceptions.RequestException as e:
            print(f"Attempt {attempt+1} failed for {url}: {str(e)}")
            if attempt < max_retries - 1:
                # Add a random delay between 5-15 seconds before retrying
                sleep_time = retry_delay + random.uniform(0, 10)
                print(f"Waiting {sleep_time:.2f} seconds before retrying...")
                time.sleep(sleep_time)


# Links of all the sites from which data needs to be extracted
urls = ["https://www.hakunamatatatech.com/our-resources/blog/iot-security-challenges",
    "https://iotsecurityfoundation.org/",
    "https://www.ftc.gov/enforcement/consumer-sentinel-network",
    "https://www.globalsign.com/en/blog/common-cyber-attacks-in-the-iot",
    "https://www.bbc.com/news/technology", 
    "https://www.theverge.com/tech",
    "https://www.csoonline.com/article/570051/iot-security-what-you-need-to-know.html",
    "https://cisomag.com/10-iot-security-incidents-that-make-you-feel-less-secure/",
    "https://www.ninjaone.com/blog/how-to-secure-iot-devices/",
    "https://www.iotforall.com/5-worst-iot-hacking-vulnerabilities",
    "https://www.wired.com/tag/iot/",
    "https://www.zdnet.com/topic/internet-of-things/",
    "https://www.techtarget.com/iotagenda/",
    "https://owasp.org/www-project-internet-of-things/",
    "https://www.forbes.com/councils/forbestechcouncil/2023/02/22/14-essential-steps-to-take-to-secure-household-iot-devices/",
    "https://securityscorecard.com/blog/internet-of-things-threats-and-risks/",
    "https://www.nist.gov/topics/internet-things-iot",
    "https://www.sentinelone.com/cybersecurity-101/data-and-ai/iot-security-risks/",
    "https://www.techtarget.com/iotagenda/",
    "https://pmc.ncbi.nlm.nih.gov/articles/PMC11014407/",
    "https://www.iotsworldcongress.com/5-infamous-iot-hacks-and-vulnerabilities/",
    "https://www.symantec.com/blogs/threat-intelligence/iot-security",
    "https://www.microsoft.com/en-us/security/business/solutions/iot-security",
    "https://aws.amazon.com/iot-device-defender/",
    "https://www.fortinet.com/resources/cyberglossary/iot-device-vulnerabilities",
    "https://www.criticalinsight.com/glossary/what-is-a-list-of-vulnerable-iot-devices",
    "https://digitalskills.continuingeducation.ncsu.edu/iot-technology-the-risks-and-rewards-of-a-connected-world/",
    "https://www.kaspersky.com/blog/riskiest-it-and-iot-devices-in-organization/51958/",
    "https://www.mdpi.com/2076-3417/10/12/4102",
    "https://www.cisa.gov/news-events/news/securing-internet-things-iot",
    "https://www.thalesgroup.com/en/markets/digital-identity-and-security/iot/magazine/internet-threats",
    "https://www.fortinet.com/resources/cyberglossary/iot-best-practices",
    "https://deviceauthority.com/security-and-privacy-issues-in-iot-generated-big-data/"]

# Extract text from each source with a delay between requests
iot_texts = {}
for url in urls:
    """
    it sends in each URL to the extract_texts function to extract texts from the particular site
    """
    print(f"Extracting text from {url}")
    text = extract_texts(url)
    iot_texts[url] = text
    # Add a delay between 3-7 seconds between requests
    delay = 3 + random.uniform(0, 4)
    print(f"Waiting {delay:.2f} seconds before next request...")
    time.sleep(delay)

"""
converts the dictionary to a DataFrame and saves it to a CSV file
"""
df_texts = pd.DataFrame(list(iot_texts.items()), columns=["Source", "Extracted_Text"])

df_texts.to_csv("iot_security_articles_test.csv", index=False)
df_texts["Extracted_Text"] = df_texts["Extracted_Text"].fillna("")
print(f"Collected text from {len(df_texts)} articles")
```

```python
def clean_text(text):
    """
    this function cleans all my text, i.e., it removes all punctuation and normalizes all text (changes all text to lowercase)

    >>> clean_text("Hello, World! How are you?")
    'hello world how are you'
    
    >>> clean_text("Test@#$%^&*()_+[]{}|;:'\",.<>?/~`")
    'test'
    
    >>> clean_text("  Multiple   spaces   and\ttabs  ")
    '  multiple   spaces   and\ttabs  '
    """
    text = re.sub(r'[^\w\s]', '', text)  # Remove punctuation
    text = text.lower()  
    return text
```

```python
"""
a list of security blogs
"""
security_blogs = [
    "https://blog.google/threat-analysis-group/",
    "https://www.microsoft.com/en-us/security/blog/",
    "https://www.welivesecurity.com/category/iot/",
    "https://blog.talosintelligence.com/",
    "https://www.mandiant.com/resources/blog",
    "https://www.crowdstrike.com/blog/category/threat-intel-research/",
    "https://www.cyberark.com/resources/blog/top-10-vulnerabilities-that-make-iot-devices-insecure",
    "https://www.balbix.com/insights/addressing-iot-security-challenges/",
    "https://www.kaspersky.com/resource-center/preemptive-safety/best-practices-for-iot-security",
    "https://www.rishabhsoft.com/blog/iot-security-challenges",
    "https://dig8ital.com/post/iot-network-security/"
]

# Collect security blog data
blog_texts = {}
for blog in security_blogs:
    """
    extracts texts from the security blogs and saves it to a dictionary
    """
    print(f"Extracting from {blog}")
    text = extract_texts(blog)
    blog_texts[blog] = text

"""
save security blog data to a DataFrame, removes all empty (none) values, and stores it in a CSV file
"""
df_blogs = pd.DataFrame(list(blog_texts.items()), columns=["Source", "Extracted_Text"])
df_blogs["Extracted_Text"] = df_blogs["Extracted_Text"].fillna("")
df_blogs.to_csv("security_blogs.csv", index=False)
```

```python
"""
listing the most common privacy breaches
"""
privacy_breaches = [
    "malware", "ransomware", "data leak", "weak authentication", "botnets", 
    "data breach", "ddos attack", "mitm attack", "unauthorized access",
    "phishing", "spyware", "zero-day exploit"
]

"""
merging all data and normalizing them
combine Reddit titles and extracted texts from articles and security blogs
"""
reddit_texts = " ".join(df_reddit["Title"].tolist())
article_texts = " ".join(df_texts["Extracted_Text"].tolist())
blog_texts_str = " ".join(df_blogs["Extracted_Text"].tolist())

# Combine all text data
combined_text = reddit_texts + " " + article_texts + " " + blog_texts_str
combined_text = combined_text.lower()
original_text = combined_text  

"""
Cleaning (removing punctuation and normalizing) the combined text
"""
combined_text = clean_text(combined_text)

breach_counts = {}
for b in privacy_breaches:
    # Count occurrences
    if " " in b:
        # multi-word phrases,counting them in the original text
        breach_counts[b] = original_text.count(b)
    else:
        breach_counts[b] = combined_text.count(b)

"""
sorts through all the breach counts to arrange all the threats in descending order
"""
sorted_breach_counts = dict(sorted(breach_counts.items(), key=lambda item: item[1], reverse=True))
print("Most Common IoT Breaches:", sorted_breach_counts)

"""
producing a visualization to see the most common IoT Security Breaches
"""
plt.figure(figsize=(16, 10))
plt.bar(list(sorted_breach_counts.keys()), list(sorted_breach_counts.values()), color='skyblue', edgecolor="black")
plt.xticks(rotation=45, ha='right')
plt.xlabel('Privacy Breach Type', fontweight='bold')
plt.ylabel('Number of Mentions', fontweight='bold')
plt.title('Most Common IoT Security Breaches', fontsize=14, fontweight='bold')
plt.tight_layout()
plt.savefig('iot_breaches.png')
plt.show()
```

```python
"""
this cell shows the results or summary statistics for my research question 1
"""
print("SUMMARY STATISTICS (RQ1):")
total_mentions = sum(sorted_breach_counts.values())
print(f"Total security issue mentions: {total_mentions}")
print(f"Top 5 most common issues: {list(sorted_breach_counts.keys())[:5]}")
```
